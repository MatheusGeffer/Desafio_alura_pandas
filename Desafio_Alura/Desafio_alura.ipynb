```python
# Dia 2
# Classificando os gêneros de acordo com a CDU - Classificação Decimal Universal.
lista_CDU = []
for cdu in emprestimos_inteiro['localizacao']:
    if (cdu < 100):
        lista_CDU.append('Generalidades')
    elif (cdu <200):
        lista_CDU.append('Filosofia e Psicologia')
    elif (cdu <300):
        lista_CDU.append('Religião')
    elif (cdu <400):
        lista_CDU.append('Ciências sociais')
    elif (cdu <500):
        lista_CDU.append('Classe vaga')
    elif (cdu <600):
        lista_CDU.append('Matemática e Ciências Naturais')
    elif (cdu <700):
        lista_CDU.append('Ciências Aplicadas')
    elif (cdu <800):
        lista_CDU.append('Belas Artes')
    elif (cdu <900):
        lista_CDU.append('Linguagem')
    else:
        lista_CDU.append('Geografia. Biografia. História.')

# Criando uma coluna com a CDU geral
emprestimos_inteiro['CDU_geral'] = lista_CDU
```

```python
# Quantidade de exemplares por ano
emprestimos_data = pd.DataFrame(emprestimos_inteiro['data_emprestimo'].value_counts()).reset_index()
emprestimos_data.columns= ['data', 'quantidade']
emprestimos_data['data'] = pd.to_datetime(emprestimos_data['data'])
emprestimos_data

```

```python
emprestimos_por_ano = emprestimos_data.groupby(by=emprestimos_data.data.dt.year)['quantidade'].sum()
emprestimos_por_ano.index.name= 'ano'
emprestimos_por_ano
```

```python
# Obter o índice (anos) e os valores (quantidades)
ano = emprestimos_por_ano.index
quantidade = emprestimos_por_ano.values

# Plotar o gráfico
plt.plot(ano, quantidade, linestyle='-', color='b')
plt.xlabel('ano')
plt.ylabel('quantidade')
plt.title('Quantidade de exemplares emprestados do SISBI por ano')
plt.grid(True)
plt.show()
```

```python
# Quantidade de exemplares por mês
emprestimos_por_mes = emprestimos_data.groupby(by=emprestimos_data.data.dt.month)['quantidade'].sum()
emprestimos_por_mes.index.name = 'mes'
emprestimos_por_mes
```

```python
# Modificar dados por os numeros por meses

dicionario_meses ={1:'jan', 2:'fev', 3:'mar', 4:'abr',
                   5:'mai', 6:'jun', 7:'jul', 8:'ago',
                   9:'set', 10:'out', 11:'nov', 12:'dez'}

emprestimos_por_mes.index = emprestimos_por_mes.index.map(dicionario_meses)
emprestimos_por_mes

```

```python
# Obter o índice (mes) e os valores (quantidades)
mes = emprestimos_por_mes.index
quantidade = emprestimos_por_mes.values

# Plotar o gráfico
plt.plot(mes, quantidade, linestyle='-', color='b')
plt.xlabel('mes')
plt.ylabel('quantidade')
plt.title('Quantidade de exemplares emprestados do SISBI por mês')
plt.suptitle('Período de 2010 a 2020')
plt.grid(True)

# Formatar os valores de y com pontos para separar as casas decimais
plt.gca().yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, p: format(int(x), ',').replace(',','.')))   
plt.show()
```

```python
# Qual o emprestimo por hora

emprestimos_por_hora = emprestimos_data.groupby(by=emprestimos_data.data.dt.hour)['quantidade'].sum()
emprestimos_por_hora.index.name = 'horario'
emprestimos_por_hora = emprestimos_por_hora.reset_index()
emprestimos_por_hora
```

```python

emprestimos_por_hora = emprestimos_por_hora.sort_values(ascending=True,by='quantidade')
```

```python
# Plotar grafico dos horarios
fig, ax = plt.subplots()
emprestimos_por_hora = emprestimos_por_hora.sort_values(by='quantidade', ascending=False)
ax.bar(emprestimos_por_hora['horario'], emprestimos_por_hora['quantidade'], color='b')
ax.set_title("Quantidade de exemplares emprestados do SISBI por faixa horária\n")
ax.text(s='Período de 2010 a 2020', x=-0.5, y=225000)
ax.set_xticks(emprestimos_por_hora['horario'])

# Formatar os números do eixo y com separador de milhar
ax.yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, p: format(int(x), ',').replace(',', '.')))
plt.show()
```

```python
# Função para analisar o percentual de frequencia

def tabela_frequencia(categoria):
 
  dataframe = pd.DataFrame(emprestimos_inteiro[categoria].value_counts())                      
  dataframe.columns = ['quantidade']
  dataframe['percentual'] = round((dataframe.quantidade / dataframe.quantidade.sum())*100,1)

  return dataframe

```

```python
# Função para gerar gráfico box plot

def gera_box_plot(dataset,x,y,titulo,subtitulo):
  

  sns.set_theme(style="darkgrid", palette='Blues',font_scale=1.3)                    
  plt.figure(figsize=(16,10))                                                           

  ax = sns.boxplot(y= y, x= x, data= dataset,color='#4171EF')                                           
  ax.yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, p: format(int(x), ',').replace(',','.')))                


  plt.ylim(0,max(dataset[y])*1.1)                                               #Definir o limite do eixo y
  plt.xlabel(None)                                                                     
  plt.ylabel(None)                                                                    

  ax.set_title(titulo+"\n",size=20,loc='left',weight='bold')
  ax.text(s=subtitulo,x=-0.5,y=max(dataset[y])*1.11,fontsize=18, ha='left',color='gray')  

  
     
```

```python
gera_box_plot(alunos_graduacao_acervo, 'ano', 'quantidade', 'Distribuição de emprestimos', 'Alunos de graduação do acervo circulante')
```
